```python
import json

with open("summary_examples.json", "r") as f:
    data = json.load(f)

example_weekly_summaries = data["example_weekly_summaries"]
```

```python
from langchain.docstore.document import Document

docs = []
for row in example_weekly_summaries:
    summary_text = row["summary"]
    contact = row["contact"]
    
    doc = Document(
        page_content=summary_text,
        metadata={
            "contact": contact,
            "week_start": row["week_start"],
            "week_end": row["week_end"],
            "created_at": row["created_at"],
        }
    )
    docs.append(doc)
```

```python
import sqlite3
from datetime import datetime, timedelta
DAYS_LOOKBACK = 30 ### How many days to look back for our query
db_path = "./messages.db"
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

def get_identifier_for_display_name(display_name: str) -> str:
    query = """
    SELECT identifier
    FROM contacts
    WHERE display_name = ?
    """
    cursor.execute(query, (display_name,))
    row = cursor.fetchone()
    if row:
        return row[0]
    return None

display_name = contact_of_interest ### ************ we are gettting the contact

identifier = get_identifier_for_display_name(display_name)
if not identifier:
    print(f"No contact found for {display_name}")
else:
    print(f"Found identifier for {display_name}: {identifier}")

# 3. Query the last XX days of messages for that identifier
if identifier:
    now = datetime.now()
    earliest_date = now - timedelta(days=DAYS_LOOKBACK)
    earliest_date_str = earliest_date.strftime("%Y-%m-%d %H:%M:%S")

    messages_query = """
    SELECT text, message_date
    FROM messages
    WHERE chat_id = ?
      AND message_date >= ?
    ORDER BY message_date ASC
    """
    cursor.execute(messages_query, (identifier, earliest_date_str))
    messages_data = cursor.fetchall()

    # messages_data is a list of tuples: [(text1, message_date1), (text2, message_date2), ...]

conn.close()
```

```python
from langchain.docstore.document import Document

docs = []
for text, msg_date in messages_data:
    print(msg_date)
    docs.append(
        Document(
            page_content=text,
            metadata={
                "message_date": msg_date,
                "contact_display_name": display_name,
                "chat_id": identifier,
            }
        )
    )
```

```python
# !pip install sentence-transformers chromadb 

from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma  

embedding_fn = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorstore = Chroma.from_documents(documents=docs, embedding=embedding_fn)
```

```python
user_query = "What topics did I discuss with Professor Loessi about my final grade?"
retrieved_docs = vectorstore.similarity_search(query=user_query, k=3)

for i, doc in enumerate(retrieved_docs):
    print(f"--- Retrieved doc #{i+1} ---")
    print(doc.metadata)
    try:
        print(doc.metadata['message_date'])
    except Exception as e:
        print(e)
    print("Text:", doc.page_content)
    print()
```

```python
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate

# user_query = "What topics did I discuss with Professor Loessi about my final grade?"


#prolly need to change this prompt
template = """
You are a helpful assistant. 
Use the following conversation messages to answer the user's query.

Messages:
{context}

Query: {query}

Answer:
"""

prompt = PromptTemplate(
    input_variables=["context", "query"],
    template=template,
)

llm = ChatGroq(
    model="mixtral-8x7b-32768",
    temperature=0,
    max_tokens=None,
    timeout=None,
    max_retries=2,
    # other params...
)

chain = LLMChain(llm=llm, prompt=prompt)

def answer_user_query(query: str):
    docs = vectorstore.similarity_search(query, k=3)
    combined_context = "\n\n".join([d.page_content for d in docs])
    return chain.run({
        "context": combined_context,
        "query": query
    })

final_answer = answer_user_query(user_query)
print("Final answer:", final_answer)
```
